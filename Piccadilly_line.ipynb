```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Create the graph
G = nx.Graph()
```

```python
# # Add edges
# , ('A', 'C', 4), ('B', 'C', 2), ('B', 'D', 2), ('B', 'E', 3),
#     ('C', 'E', 3), ('D', 'F', 1), ('E', 'F', 3)
edges = [
    ('A', 'B', 1),
    ('B', 'C', 2),
    ('C', 'D', 3),
    ('D', 'E', 3),
    ('E', 'F', 3),

]

G.add_weighted_edges_from(edges)
```

```python
# Define positions for a visually appealing layout
 # 'A': (0, 2),
    # 'D': (2, 2),
    # 'E': (2, 0),
    # 'F': (3, 1)
pos = {
    # (vertical, horizontal)
        'A': (1, 0),
        'B': (2, 1),
        'C': (3, 1),
        'D': (4, 1),
        'E': (5, 2),
        'F': (6, 3),
}
```

```python
# NOTE: - change color of edge and node,
# Need labels for nodes
# need a key table


edge_labels = {(u, v): d for u, v, d in edges}
# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=700, node_color='blue', font_size=14, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=14)

visited_nodes = ['A', 'B', 'D', 'C', "E", "F"]
nx.draw_networkx_nodes(G, pos, nodelist=visited_nodes, node_color='blue')

plt.title('Initial Map with Distances')
plt.show()
```

```python

import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

edge_labels={
    #piccadilly
    ('A', 'B'): '1.0km',
    ('B', 'C'): '0.48km',
    ('C', 'D'): '0.6km',
    ('D', 'E'): '0.26km',
    ('E', 'F'): '0.52km',
    #victoria
    ('G', 'H'): '1.6km',
    ('H', 'B'): '1.2km',
    ('B', 'J'): '1.0km',
    ('J', 'K'): '1.2km',
    ('K', 'L'): '0.4km',
    #central
    ('M', 'N'): '1.2km',
    ('N', 'O'): '0.6km',
    ('O', 'J'): '0.6km',
    ('J', 'P'): '0.7km',
    ('P', 'F'): '1.3km',
    #Jubilee 
    ('Q', 'R'): '0.7km',
    ('R', 'S'): '1.4km',
    ('S', 'T'): '1.6km',
    ('T', 'O'): '1.6km',
    ('O', 'B'): '1.3km',
    #Bakerloo 
    ('U', 'T'): '0.2km',
    ('T', 'V'): '0.9km',
    ('V', 'J'): '1.0km',
    ('J', 'C'): '0.8km',
    ('C', 'W'): '0.6km',
}

# Create a graph object
MyGraph = nx.Graph()

# Add nodes
# Piccadilly
MyGraph.add_node('A', npos=(3.5, -6), ccn='#0000FF')
MyGraph.add_node('B', npos=(4.5,0), ccn='white', node_size=110)  # 'B': 'Green Park'
MyGraph.add_node('C', npos=(5.2, 0), ccn='#0000FF')
MyGraph.add_node('D', npos=(6.1, 0), ccn='#0000FF')
MyGraph.add_node('E', npos=(6.6, 2.8), ccn='#0000FF')
MyGraph.add_node('F', npos=(7.3, 6.8), ccn='#0000FF')

# Victoria
MyGraph.add_node('G', npos=(4.5, -18), ccn='#3CC5E7')
MyGraph.add_node('H', npos=(4.5, -8.5), ccn='#3CC5E7')
MyGraph.add_node('J', npos=(5, 8), ccn='#3CC5E7')
MyGraph.add_node('K', npos=(6.1, 14.5), ccn='#3CC5E7')
MyGraph.add_node('L', npos=(7, 18.5), ccn='#3CC5E7')

# Central
MyGraph.add_node('M', npos=(2, -2), ccn='#ff0000')
MyGraph.add_node('N', npos=(3, 3.5), ccn='#ff0000')
MyGraph.add_node('O', npos=(3.8, 8), ccn='#ff0000')
MyGraph.add_node('P', npos=(6.1, 7.5), ccn='#ff0000')

# Jubilee
MyGraph.add_node('Q', npos=(2.2, 41), ccn='#616161')
MyGraph.add_node('R', npos=(2.5, 35), ccn='#616161')
MyGraph.add_node('S', npos=(3, 28), ccn='#616161')
MyGraph.add_node('T', npos=(3.5, 21), ccn='#616161')

# Bakerloo
MyGraph.add_node('U', npos=(3, 21), ccn='#78250B')
MyGraph.add_node('V', npos=(4.1, 15), ccn='#78250B')
MyGraph.add_node('W', npos=(6.1, -7), ccn='#78250B')

# Connect the Piccadilly
# Piccadilly
MyGraph.add_edge('A', 'B', cce='#0000FF')
MyGraph.add_edge('B', 'C', cce='#0000FF')
MyGraph.add_edge('C', 'D', cce='#0000FF')
MyGraph.add_edge('D', 'E', cce='#0000FF')
MyGraph.add_edge('E', 'F', cce='#0000FF')
# Victoria
MyGraph.add_edge('G', 'H', cce='#3CC5E7')
MyGraph.add_edge('H', 'B', cce='#3CC5E7')
MyGraph.add_edge('B', 'J', cce='#3CC5E7')
MyGraph.add_edge('J', 'K', cce='#3CC5E7')
MyGraph.add_edge('K', 'L', cce='#3CC5E7')
# Central
MyGraph.add_edge('M', 'N', cce='#ff0000')
MyGraph.add_edge('N', 'O', cce='#ff0000')
MyGraph.add_edge('O', 'J', cce='#ff0000')
MyGraph.add_edge('J', 'P', cce='#ff0000')
MyGraph.add_edge('P', 'F', cce='#ff0000')
# Jubilee 
MyGraph.add_edge('Q', 'R', cce='#616161')
MyGraph.add_edge('R', 'S', cce='#616161')
MyGraph.add_edge('S', 'T', cce='#616161')
MyGraph.add_edge('T', 'O', cce='#616161')
MyGraph.add_edge('O', 'B', cce='#616161')
# Bakerloo
MyGraph.add_edge('U', 'T', cce='#78250B')
MyGraph.add_edge('T', 'V', cce='#78250B')
MyGraph.add_edge('V', 'J', cce='#78250B')
MyGraph.add_edge('J', 'C', cce='#78250B')
MyGraph.add_edge('C', 'W', cce='#78250B')

# Extract attributes from the graph to dictionaries
pos = nx.get_node_attributes(MyGraph, 'npos')
nodecolour = nx.get_node_attributes(MyGraph, 'ccn')
edgecolour = nx.get_edge_attributes(MyGraph, 'cce')

# Place the dictionary values in array
nodearray = nodecolour.values()
edgearray = edgecolour.values()

# Define graph size
plt.figure(figsize=(20, 18))

# Display the name of the stations
# Piccadilly
plt.text(pos['A'][0]-0.1, pos['A'][1]-1.3 , s='Hyde Park Corner', rotation=0)
plt.text(pos['B'][0] + 0.1, pos['B'][1] +0.1 , s='Green Park', rotation=15)
plt.text(pos['C'][0] + 0.1, pos['C'][1] + 0.1, s='Piccadilly Circus', rotation=15)
plt.text(pos['D'][0] - 0.4, pos['D'][1] - 1, s='Leicester Square', rotation=0)
plt.text(pos['E'][0], pos['E'][1]-1  , s='Covent Garden', rotation=0)
plt.text(pos['F'][0] , pos['F'][1]+1 , s='Holborn', rotation=0)
# Victoria
plt.text(pos['G'][0]-0.3, pos['G'][1]-1 , s='Pimlico', rotation=0)
plt.text(pos['H'][0]+0.2, pos['H'][1]-0.1 , s='Victoria', rotation=0)
plt.text(pos['J'][0]+0.1, pos['J'][1]+0.8 , s='Oxford Circus', rotation=45)
plt.text(pos['K'][0]-0.5, pos['K'][1]+0.8 , s='Warren Street', rotation=0)
plt.text(pos['L'][0]-0.2, pos['L'][1]+0.8 , s='Euston', rotation=0)
# central
plt.text(pos['M'][0]-0.5, pos['M'][1]-1 , s='Lancaster Gate', rotation=0)
plt.text(pos['N'][0], pos['N'][1]-1.4 , s='Marble Arch', rotation=0)
plt.text(pos['O'][0]+0.1, pos['O'][1]+0.8 , s='Bond Street', rotation=45)
plt.text(pos['P'][0]-0.5, pos['P'][1]+1 , s='Tottenham Court Road', rotation=0)
# jubilee 
plt.text(pos['Q'][0]-0.3, pos['Q'][1]+1, s='Finchley Road', rotation=0)
plt.text(pos['R'][0]+0.2, pos['R'][1]-0.1 , s='Swiss Cottage', rotation=0)
plt.text(pos['S'][0]+0.2, pos['S'][1]-0.1 , s='St John\'s Wood', rotation=0)
plt.text(pos['T'][0]+0.2, pos['T'][1]+0.4 , s='Baker Street', rotation=30)
# bakerloo
plt.text(pos['U'][0]-0.3, pos['U'][1]+0.8 , s='Marylebone', rotation=0)
plt.text(pos['V'][0]+0.1, pos['V'][1]+0.5 , s='Regent\'s Park', rotation=0)
plt.text(pos['W'][0]-0.2, pos['W'][1]-1.3 , s='Charing Cross', rotation=0)

# Define types of node
special_nodes = ['B', 'J', 'F', 'O', 'T', 'J', 'C']
normal_nodes = [node for node in MyGraph.nodes if node not in special_nodes ]
# Get the colors of the normal nodes
normal_colors = [nodecolour[node] for node in normal_nodes]

# Draw the graph's nodes and edges
# Draw the nodes
nx.draw_networkx_nodes(MyGraph, pos, nodelist= normal_nodes, node_color=normal_colors,node_size=500, edgecolors='black')

# Draw the special node 'B' with different attributes
nx.draw_networkx_nodes(MyGraph, pos, nodelist=special_nodes, node_color='white', node_size=700, edgecolors='black')

# nx.draw_networkx(MyGraph, pos, node_color=nodearray, with_labels=False)
nx.draw_networkx_edges(MyGraph, pos, edge_color=edgearray, width=5)
nx.draw_networkx_edge_labels(
    MyGraph, pos,
    edge_labels=edge_labels,
    font_color='black',
    font_weight='bold'
)


# Create a line for the legend
piccadilly_line = mlines.Line2D([], [], color='#0000FF', label='Piccadilly')
victoria_line = mlines.Line2D([], [], color='#3CC5E7', label='Victoria')
central_line = mlines.Line2D([], [], color='#ff0000', label='Central')
jubilee_line = mlines.Line2D([], [], color='#616161', label='Jubilee ')
bakerloo_line = mlines.Line2D([], [], color='#78250B', label='Bakerloo')
interchangable = mlines.Line2D([], [], marker='o', color='black', markerfacecolor='white',
                               markersize=15, linestyle='None', label='Interchangable Station')

# Add the line to the legend
first_legend = plt.legend(handles=[piccadilly_line, victoria_line, central_line, jubilee_line, bakerloo_line, interchangable], loc='lower right', fancybox=True, framealpha=1, shadow=True, borderpad=1, title="Key")

# Manually add the first legend back to the plot
plt.gca().add_artist(first_legend)

# Title of the graph
plt.title('London Tube Line', fontsize = 40)

# Visualise the graph
plt.show()
```

```python
import numpy as np
"""
put all the length between station into a numpy array
"""
# Lengths of edges in kilometers
lengths = np.array([
    1.0, 0.48, 0.6, 0.26, 0.52,  # Piccadilly
    1.6, 1.2, 1.0, 1.2, 0.4,     # Victoria
    1.2, 0.6, 0.6, 0.7, 1.3,     # Central
    0.7, 1.4, 1.6, 1.6, 1.3,     # Jubilee
    0.2, 0.9, 1.0, 0.8, 0.6      # Bakerloo
])

# Total length of transportation network
total_length = sum(lengths)


# THe average distance between stations
avg_distance = np.mean(lengths)

# standard deviation of the distances between the stations
std_length = np.std(lengths)

print(f"Total Length: {total_length}\nAverage Distance: {avg_distance}\nStandard Deviation Distance: {std_length}")
```
